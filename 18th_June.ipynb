1. What is the role of the 'else' block in a try-except statement? Provide an example
scenario where it would be useful.

**Answer:** The 'else' block in a try-except statement is used when we want to execute a set of code only if there has been no exception caught in try block.

**Else** block indicates that the piece of code in try block has been executed successfully without any issues and have reached the section where the code will not throw any error. Else bloack is an optional block of code.


```python
#In this case we are checking whether divide by 0 is possible if yes then only we shall print the result else we shall print the exception.
x=10
try:
    result = x / 0
except Exception as e:
    print(e)
else:
    print(f"Result={result}")


x=100
try:
    result = x / 10
except ZeroDivisionError:
    print("ZeroDivisionError | Denominator should not be equal to 0.")
else:
    print(f"Result={result}")
```

2. Can a try-except block be nested inside another try-except block? Explain with an
example.

**Answer:** Yes, we can have a try-except block nested inside another try-except block.

```python
#if the numbers are even then need to multiply by 10 and return a list.
def demo(list_a):
  list_b = []
  try:
    for i in list_a:
        try:
          if(i%2==0):
            list_b.append(i*10)
            print("List b has :" + list_b)
        except Exception as e:
          print("Exception caught in second Try Block : " + str(e))
          break
  except Exception as e:
    print("Exception caught in first Try Block : " + str(e))
```

```python
#exception caught because the varible passed is int whereas the expected is list.
demo(1)
```

```python
#exception caught because concat between list and string cannot happen.
demo([0,1,2,3,4])
```

3. How can you create a custom exception class in Python? Provide an example that
demonstrates its usage.

**Answer:** We can create a custom exception class by declaring a class with the type of message we want to display to the users for certain type of exceptions so that in case if the specified type of error occurs we shall call that class and the required message will be displayed to the user or other action to be performed shall be carried out.

Below is an example for the same

```python
class IncorrectAge(Exception):
    def __init__(self, message):
          self.message = message

def CheckAge(a,b):
  try:
    if(a<b):
      raise IncorrectAge("Father's age is greater than son's age.")
    else:
      print("Ages entered are correct.")
  except IncorrectAge as e:
    print('A New Exception occurred: ', e.message)

CheckAge(40,60)
CheckAge(40,15)
```

10. Write a Python program that logs a message to a file named "app.log" with the
following requirements:


● The log message should be "Hello, World!"

● The log level should be set to "INFO."

● The log file should append new log entries without overwriting previous ones.

```python
import os
import logging

logger = logging.getLogger()
logger.setLevel(logging.INFO)

if os.path.isfile(r'/content/app.log'):
    print('File Exists')
    os.remove(r'/content/app.log')
    handler = logging.FileHandler(r'/content/app.log')
    logger.addHandler(handler)
    logger.info('Hello World!')
else:
    print('File Dont Exists')
    handler = logging.FileHandler(r'/content/app.log')
    logger.addHandler(handler)
    logger.info('Hello World!')
```
